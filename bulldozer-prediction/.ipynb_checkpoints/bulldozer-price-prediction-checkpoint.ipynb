### 4. Modelling

**Model driven EDA**

```python
## Because we are using Train and validation test case differently we have to write a function to set up our dataset

def processData(df):
    """
    Processing a data to the best way to use to our prediction
    
    df: A pandas DataFrame
    
    """
    # changin the saledate to saleYear,saleMonth and saleDay

    df['saleYear'] = df.saledate.dt.year
    df['saleMonth'] = df.saledate.dt.month
    df['saleDay'] = df.saledate.dt.day
    df['saleDayOfWeek'] = df.saledate.dt.dayofweek
    df['saleDayOfYear'] = df.saledate.dt.dayofyear
    # now we change the sale date now we remove the saledate column because we dont need it anymore
    df.drop('saledate',axis=1,inplace=True)
    
    # lets change the object columns to categorical column

    # To make the work easy we loop over the columns and fill it

    for name,column in df.items():
        if not pd.api.types.is_numeric_dtype(column):
            df[name+'_is_null'] = column.isna()
            df[name] = pd.Categorical(column).codes+1
        # lets fill a missing value of integer columns with median of their value
        # To make the work easy we loop over the columns and fill it
        if pd.api.types.is_numeric_dtype(column):
            df[name+'_is_null'] = column.isna()
            df[name].fillna(column.median(),inplace=True)
    return df
```

```python
bulldozers = processData(bulldozers)

bulldozers_test = processData(bulldozers_test)
```

### now we split our data into training and validation set accourding to the documentation of the client 
**Train data upto the end of 2011**

**Validation data 2012**


```python
bulldozers.saleYear
```

```python
bulldozers.saleYear.value_counts()
```

```python
train = bulldozers[bulldozers.saleYear != 2012]
valid = bulldozers[bulldozers.saleYear == 2012]

x_train, y_train = train.drop('SalePrice',axis=1), train.SalePrice
x_valid, y_valid = valid.drop('SalePrice',axis=1), valid.SalePrice
```

```python
x_train.shape,y_train.shape,x_valid.shape,y_valid.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_squared_log_error,mean_absolute_error
```

```python
%%time

np.random.seed(42)
model = RandomForestRegressor(n_jobs=-1)
model.fit(x_train,y_train)
```

```python
## all regression scoring method an one 

def regScore(model):
    
    y_pred = model.predict(x_valid)
    
    print(f"MSE {mean_squared_error(y_pred,y_valid)}")
    print(f"MAE {mean_absolute_error(y_pred,y_valid)}")
    print(f"r2 {model.score(x_valid,y_valid)}")
    print(f"RMSLE {np.sqrt(mean_squared_log_error(y_pred,y_valid))}")
```

```python
regScore(model)
```

### Now we get the baseline model result next we have to tune our hyperparameter using `RandomizedSearchCV`

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
```

```python
%%time
rs_grid = {
    'n_estimators':np.arange(10,1000,50),
    'max_depth': [None,2,6,8],
    'min_samples_split':np.arange(2,20,2),
    'min_samples_leaf':np.arange(1,20,2),
    'max_features': [1,None,'sqrt','auto'],
    'max_samples':[20000]
}

np.random.seed(42)
rs_model = RandomizedSearchCV(RandomForestRegressor(n_jobs=-1),rs_grid,n_iter=20,cv=5,verbose=True)


rs_model.fit(x_train,y_train)
```
